In this project we will use SQLite to create a database, query that database, and assign the results of that query to a pandas dataframe.

First we will import sqlite3 and pandas:

```python
import sqlite3
import pandas as pd
```

Next we will create a new database called "Chinook" and run a sql script to populate it:

```python
con = sqlite3.connect('Chinook.db')
with con:
    cur = con.cursor()
    with open('C:/Users/cscam/Chinook_Sqlite.sql', 'r', encoding='utf8') as f:
        cur.executescript(f.read())
```

Now that we have our database populated with tables, we can write a query to extract the data we need, specifically customer last names, first names, purchased track names, and associated album titles.

We will then assign the results to a DataFrame object. 

```python
cur.execute("""SELECT LastName, FirstName, track.name, album.title FROM customer
            INNER JOIN invoice ON invoice.CustomerId = customer.CustomerId
            INNER JOIN invoiceline ON invoiceline.InvoiceId = invoice.invoiceId
            INNER JOIN track ON track.TrackId = invoiceline.TrackId
            INNER JOIN album ON album.AlbumId = track.AlbumId;""")

df = pd.DataFrame(cur.fetchall(),
                  columns=['Last Name', 'First Name', 'Track Name', 'Album Title'])
df
```

Finally, we will sort the DataFrame by customer last name and first name:

```python
df.sort_values(['Last Name', 'First Name'])
```
